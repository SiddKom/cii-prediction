```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv('cii_simulation_log.csv')

# Convert 'Time' to datetime if it's not already
df['Time'] = pd.to_datetime(df['Time'])

# Plot time vs predicted CII
plt.figure(figsize=(12, 6))
plt.plot(df['Time'], df['Predicted_CII'], label='Predicted CII', color='green', linewidth=2, marker='o')

# Format the plot
plt.xlabel('Time', fontsize=12)
plt.ylabel('Predicted CII', fontsize=12)
plt.title('Predicted CII Over Time', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
